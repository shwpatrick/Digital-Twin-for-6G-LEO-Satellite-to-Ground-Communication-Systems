```python
def plot_bar_satellite_location_samples(df) :
    
    # folder
    # ===================================================
    if not os.path.exists('data/FigurePlot/bar_satellite_location_samples' ):
        os.makedirs( 'data/FigurePlot/bar_satellite_location_samples' )
        print('create directory : bar_satellite_location_samples' )

    folder_path = 'data/FigurePlot/bar_satellite_location_samples/'
    
    # fliter
    # ===================================================
    fliter_LOS  = (df['State']=='LOS')
    fliter_NLOS = (df['State']=='NLOS')
    
    buffer_df = df
    buffer_df = buffer_df[(buffer_df['SAT_EL']<high_threshold)&(buffer_df['SAT_EL']>low_threshold)]
    
    location_list = df["location"].unique()
    for location in location_list :
    
        df = buffer_df[buffer_df['location']==location]
        
        print(location)
        print(df.shape[0])
        
        # calclulate
        # ===================================================
        df_LOS = df[fliter_LOS]
        df_NLOS = df[fliter_NLOS]
        df_total = df[fliter_LOS | fliter_NLOS]

        LOS_SAT_list = df_LOS["SAT_name"].unique()
        NLOS_SAT_list = df_NLOS["SAT_name"].unique()
        total_SAT_list = df_total["SAT_name"].unique()


        LOS_list_name  = []
        LOS_list_count = []
        NLOS_list_name  = []
        NLOS_list_count = []
        total_list_name  = []
        total_list_count = []

        for sat_name in total_SAT_list :

            total_list_name.append(sat_name)

            LOS_list_count.append(df_LOS[df_LOS['SAT_name']==sat_name].shape[0])
            NLOS_list_count.append(df_NLOS[df_NLOS['SAT_name']==sat_name].shape[0])
            total_list_count.append(df_total[df_total['SAT_name']==sat_name].shape[0])

        # print info
        # ===================================================
        # print('file :',name)
        # print('LOS/NLOS :',str(df_LOS.shape[0]),'/',str(df_NLOS.shape[0]))
        # print('LOS(%)',str(df_LOS.shape[0]*100/(df_LOS.shape[0]+df_NLOS.shape[0])))

        # plot figure
        # ===================================================
        name_idx = np.arange(len(total_list_name))
        width = 0.3


        fig, ax = plt.subplots(figsize=(20, 10))

        
        percentage_NLOS_list =  [a*100 / b for a,b in zip(NLOS_list_count, total_list_count)]
        percentage_LOS_list =  [a*100 / b for a,b in zip(LOS_list_count, total_list_count)]
        
        #ax.bar(name_idx,total_list_count,color='b',width=0.6, label = "Total")# 第二組數據置中對齊
        #ax.bar(name_idx-0.5*width,NLOS_list_count,color='r',width=0.3, label = "NLOS")# 第二組數據置中對齊
        #ax.bar(name_idx+0.5*width,LOS_list_count,color='g',width=0.3, label = "LOS")  # 第一組數據靠左邊緣對齊
        ax.bar(name_idx-0.5*width,percentage_NLOS_list,color='r',width=0.3, label = "NLOS")# 第二組數據置中對齊
        ax.bar(name_idx+0.5*width,percentage_LOS_list,color='g',width=0.3, label = "LOS")  # 第一組數據靠左邊緣對齊
        
        # ticks = ax.set_xticklabels([''] + total_list_name)
        # for t in ticks:
        #     t.set_rotation(45)

        #plt.xticks(rotation=70)


        #name_split = name.split('_')
        #plt.title('LOS/NLOS percentage for every 10 elevation degrees, date : ' + name_split[3] + ', location : ' + name_split[2] + ', satellite amount : ' + name_split[4], fontsize = 24)
        plt.xlabel('Satellite name', fontsize = 24)
        plt.ylabel('Percentage (%)', fontsize = 24)
        plt.ylim(0,100)

        ax.set_xticks(name_idx)
        ax.set_xticklabels(total_list_name, rotation=65)
        ax.tick_params(axis='both', which='major', labelsize=20)

        fig.subplots_adjust(bottom=0.3)
        fig.tight_layout()

        plt.legend(loc='upper left', prop={'size': 24})

        #plt.show()  
        plt.savefig(folder_path + 'bar_satellite_' + location + '_' + str(low_threshold) + '_to_' + str(high_threshold) + '.png')
        plt.close()
```
