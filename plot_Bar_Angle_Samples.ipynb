```python
def plot_bar_angle_samples(df):

    # folder
    # ===================================================
    if not os.path.exists( 'data/FigurePlot/bar_angle_samples' ):
        os.makedirs( 'data/FigurePlot/bar_angle_samples' )
        print('create directory : bar_angle_samples' )

    folder_path = 'data/FigurePlot/bar_angle_samples/'
    
    # fliter
    # ===================================================
    fliter_LOS  = (df['State']=='LOS')
    fliter_NLOS = (df['State']=='NLOS') 
    
    
    buffer_df = df
    
    location_list = df["location"].unique()
    for location in location_list :
    
        df = buffer_df[buffer_df['location']==location]
    
        print(location)
        print(df.shape[0])
    
        # calculation
        # ===================================================
        bar_LOS_num  = []
        bar_ALL_num  = []
        bar_NLOS_num  = []
        bar_name = []

        for i in range (0,9) :

            fliter_angle = ((i*10<df['SAT_EL']) & (df['SAT_EL']<=i*10+10))

            LOS_num = df[fliter_angle][fliter_LOS].shape[0]
            NLOS_num = df[fliter_angle][fliter_NLOS].shape[0]
            total_num = LOS_num + NLOS_num


            #print(i*10,'< angle <=',i*10+10,' percentage: ',percentage)


            bar_LOS_num.append(LOS_num)
            bar_NLOS_num.append(NLOS_num)
            bar_ALL_num.append(total_num)

            degree_sign = u'\N{DEGREE SIGN}'
            bar_name.append(str(i*10)+degree_sign+' < angle <= '+(str(i*10+10))+degree_sign)

        # plot
        # ===================================================
        name_idx = np.arange(len(bar_name))
        width = 0.3


        fig, ax = plt.subplots(figsize=(20, 10))

        # name_split = name.split('_')
        # plt.title('Samples for every 10 elevation degrees, date : ' + name_split[3] + ', location : ' + name_split[2] + ', satellite amount : ' + name_split[4], fontsize = 24)
        plt.xlabel('Elevation angle interval (degree)', fontsize = 30)
        plt.ylabel('Samples', fontsize = 30)
        plt.ylim(0,700)

        ax.bar(name_idx,bar_ALL_num,color='b',width=0.6, label = "Total")# 第二組數據置中對齊
        ax.bar(name_idx-0.5*width,bar_NLOS_num,color='r',width=0.3, label = "NLOS")# 第二組數據置中對齊
        ax.bar(name_idx+0.5*width,bar_LOS_num,color='g',width=0.3, label = "LOS")  # 第一組數據靠左邊緣對齊

        # plt.plot( bar_name, bar_ALL_num , color = 'blue' , linewidth = 2, marker = '*', markersize = 18, label = 'Total')
        # plt.plot( bar_name, bar_NLOS_num, color = 'red'  , linewidth = 2, marker = 'v', markersize = 18, label = 'NLOS' )
        # plt.plot( bar_name, bar_LOS_num , color = 'green', linewidth = 2, marker = '^', markersize = 18, label = 'LOS' )



        plt.legend(loc='upper right', prop={'size': 30})
        ax.set_xticks(name_idx)
        ax.set_xticklabels(bar_name, rotation=45)
        ax.tick_params(axis='both', which='major', labelsize=30)

        fig.subplots_adjust(bottom=0.3)
        fig.tight_layout()

        plt.savefig(folder_path + 'bar_angle_samples_' + location + '.png')
        #plt.show()
        plt.close()
```
