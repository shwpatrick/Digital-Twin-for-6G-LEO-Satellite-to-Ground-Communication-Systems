```python
def plot_bar_angle_percentage(df):

    # folder
    # ===================================================
    if not os.path.exists('data/FigurePlot/bar_angle_percentage' ):
        os.makedirs( 'data/FigurePlot/bar_angle_percentage' )
        print('create directory : bar_angle_percentage' )

    folder_path = 'data/FigurePlot/bar_angle_percentage/'
    
    
    
    fliter_LOS  = (df['State']=='LOS')
    fliter_NLOS = (df['State']=='NLOS') 
    
    buffer_df = df
    
    location_list = df["location"].unique()
    for location in location_list :
    
        df = buffer_df[buffer_df['location']==location]
    
        bar_num_LOS  = []
        bar_num_NLOS = []
        bar_name = []

        for i in range (0,9) :

            fliter_angle = ((i*10<df['SAT_EL']) & (df['SAT_EL']<=i*10+10))

            LOS_num = df[fliter_angle][fliter_LOS].shape[0]
            NLOS_num = df[fliter_angle][fliter_NLOS].shape[0]
            total_num = LOS_num + NLOS_num

            try : 
                LOS_percentage = LOS_num*100/total_num 
            except :
                LOS_percentage = 0

            try : 
                NLOS_percentage = NLOS_num*100/total_num 
            except :
                NLOS_percentage = 0

            #print(i*10,'< angle <=',i*10+10,' percentage: ',percentage)


            bar_num_LOS.append(LOS_percentage)
            bar_num_NLOS.append(NLOS_percentage)

            degree_sign = u'\N{DEGREE SIGN}'
            bar_name.append(str(i*10)+degree_sign+' < angle <= '+(str(i*10+10))+degree_sign)


        fig, ax = plt.subplots(figsize=(20, 10))
        name_idx = np.arange(len(bar_name))
        width = 0.4

        # plt.bar(bar_name,bar_num_LOS)
        # plt.plot(bar_name,bar_num_NLOS,color='red')
        # plt.plot(bar_name,bar_num_LOS,color='green')


        # plt.plot( bar_name, bar_num_NLOS, color = 'red'  , linewidth = 2, marker = 'v', markersize = 18, label = 'NLOS' )
        # plt.plot( bar_name, bar_num_LOS , color = 'green', linewidth = 2, marker = '^', markersize = 18, label = 'LOS' )    



        ax.bar(name_idx-0.5*width,bar_num_NLOS,color='r',width=0.4, label = "NLOS")# 第二組數據置中對齊
        ax.bar(name_idx+0.5*width,bar_num_LOS,color='g' ,width=0.4, label = "LOS")  # 第一組數據靠左邊緣對齊


        plt.xticks(rotation=70)

        # name_split = name.split('_')
        # plt.title('Samples for every satellite, date : ' + name_split[3] + ', location : ' + name_split[2] + ', satellite amount : ' + name_split[4], fontsize = 24)
        plt.xlabel('Elevation angle interval (degree)', fontsize = 30)
        plt.ylabel('LOS/NLOS percentage (%)', fontsize = 30)

        ax.set_xticks(name_idx)
        ax.set_xticklabels(bar_name, rotation=65)
        ax.tick_params(axis='both', which='major', labelsize=30)

        fig.subplots_adjust(bottom=0.3)
        fig.tight_layout()

        plt.legend(loc='upper left', prop={'size': 30})


        plt.savefig(folder_path + 'bar_angle_percentage_' + location + '.png')
        plt.close()
```
